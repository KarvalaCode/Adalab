## Ejercicios Unión de Datos  

Para realizar este pair programming deberéis usar el conjunto de datos de world-data-2023-part1.csv y el de world-data-2023-part2.csv.  

Hoy tenemos un dataset nuevo, en este encontraremos las siguientes columnas:  

- **GDP**: Producto Interno Bruto, el valor total de bienes y servicios producidos en el país.  
- **Gross primary education enrollment** (%): Tasa bruta de matriculación en educación primaria.
- **Gross tertiary education enrollmen**t (%): Tasa bruta de matriculación en educación terciaria.

- **Official language**: Idioma(s) oficial(es) hablado(s) en el país.

- **Tax revenue** (%): Ingresos fiscales como porcentaje del PIB.
- **Total tax rate**: Carga fiscal total como porcentaje de las ganancias comerciales.
- **Unemployment rate**: Porcentaje de la fuerza laboral que está desempleada.
- **Urban_population**: Porcentaje de la población que vive en áreas urbanas.
- **coordinates**: Coordenadas de latitud y longitud de la ubicación del país.
- **country**: Nombre del país.  

Tienes a tu disposición dos conjuntos de datos, "world-data-2023-part1.csv" y "world-data-2023-part2.csv", que contienen información de una serie de indicadores y datos de distintos países. ## Ejercicios de Limpieza

1. Después de la unión de datos, tenemos dos columnas de "country". Elimina una de ellas.

```python
world
```

```python
world.drop('country', axis=1, inplace=True)
```

```python
world
```

2. Los nombres de las columnas no son homogeneos. Cambia los nombres de las columnas de tal forma que:  
- No tengan espacios.  
- Estén en minúscula.  
- No tengan paréntesis, es decir, quitar "(%)", "(Km2)".
- Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas.
- Algunas columnas tienen ":". Eliminalos de los nombres de las columnas.

```python
# - No tengan espacios.  
# - Estén en minúscula. 
# - No tengan paréntesis, es decir, quitar "(%)", "(Km2)".
# - Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas.
# - Algunas columnas tienen ":". Eliminalos de los nombres de las columnas.

cambios = {columna: columna.lower().replace(' ', '').replace('(%)', '').replace('(km2)', '').replace('\n', '').replace(':', '').replace('(p/km2)', '') for columna in world.columns}

world.rename(columns = cambios, inplace = True)

world
```

3. La columnas coordinates tiene la latitud y la longitud en una sola columna. Crea dos columnas nuevas, una con la longitud y otra con la latitud. Una vez hecho, elimina la columna de coordinates.

```python
long_lat = world['coordinates'].str.split("[),(]", expand = True)
world[['longitude', 'latitude']] = long_lat.get([1,2])
world
```

```python
world.drop('coordinates', axis=1, inplace=True)
world
```

4. Las columnas:
- unemployment_rate, total_tax_rate , tax_revenue ,
population_labor_force_participation , out_of_pocket_health_expenditure ,
gross_tertiary_education_enrollment , gross_primary_education_enrollment ,
forested_area , cpi_change , agricultural_land tienen "%". Elimina los "%" de los valores de las columnas.

```python
world.loc[:, ['unemploymentrate','totaltaxrate','taxrevenue','populationlaborforceparticipation','outofpockethealthexpenditure','grosstertiaryeducationenrollment','grossprimaryeducationenrollment','forestedarea','cpichange','agriculturalland']]
```

```python
porcentage = ['unemploymentrate','totaltaxrate','taxrevenue','populationlaborforceparticipation','outofpockethealthexpenditure','grosstertiaryeducationenrollment','grossprimaryeducationenrollment','forestedarea','cpichange','agriculturalland']

for columna in porcentage:
    world[columna] = world[columna].str.replace('%', '')
```

5. Haz lo mismo para las columnas de gasoline_price , gdp , minimum_wage , pero eliminando "$".

```python
dolar = ['gasolineprice','gdp','minimumwage']

for columna in dolar:
    world[columna] = world[columna].str.replace('$', '')
```

## Ejercicios de Filtrado

1. Encuentra todos los países cuya mortalidad infantil esté entre 40 y 50 personas por kilómetro cuadrado.

```python
world.columns
```

```python
mortalidad_40_50 = world['infantmortality'].between(40,50, inclusive='both')
world[mortalidad_40_50]
```

2. Encuentra los países cuyas tasas de natalidad son mayores o iguales a 20 y su esperanza de vida es mayor de 75 años.

```python
natalidad_20 = world['birthrate'] >= 20
esperanza_75 = world['lifeexpectancy'] > 75

world[natalidad_20 & esperanza_75]
```

3. Encuentra las ciudades cuyos paises contienen la palabra "la" en su nombre.

```python
pais_la = world['country'].str.contains('la', case=True, na=True, regex=True)

world[pais_la][['capital/majorcity', 'largestcity']]
```

```python
euro = world['currency-code'] == 'EUR'
natalidad = world['birthrate'] > world['birthrate'].mean()

world[euro & natalidad]['country']
```
