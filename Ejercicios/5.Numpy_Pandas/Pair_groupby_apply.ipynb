```python
def to_float(columna):
    try:
        return float(columna.replace("'","").replace(',',''))
    except:
        return np.nan
```

```python
lista_col=['landarea','armedforcessize', 'density', 'co2-emissions', 'gdp', 'population', 'urban_population', 'longitude', 'latitude', 'cpi']

for columna in lista_col:
    world[columna]=world[columna].apply(to_float)
```

2. Crea una función para añadir una nueva columna llamada 'Continent' al DataFrame para indicar a qué continente pertenece cada país.

```python
world.rename(columns={'longitude':'latitude','latitude':'longitude'},inplace=True)
```

```python
def continent (latitud,longitud):
    if 35.59<=latitud<=71.1 and -9.26<=longitud<=40.14:
        return 'Europa'
    elif -10<=latitud<=80 and -30<=longitud<=180:
        return 'Asia'
    elif -55<=latitud<=90 and -169.4<=longitud<=-26.4:
        return 'America'
    elif -35<=latitud<=35 and -20<=longitud<=50:
        return 'Africa'
    else:
        return 'Oceania'
    
```

```python
world['Continent']=world.apply(lambda dataframe: continent (dataframe['latitude'],dataframe['longitude']), axis=1)
```

```python
world[['Continent','country']]
```

```python
densidad_tax=tasa_impuestos.merge(densidad, how='inner', on='Continent')
```

```python
def tax_person(tax,density):
    return tax/density
```

```python
densidad_tax['tax_person']=densidad_tax.apply(lambda dataframe: tax_person (dataframe['totaltaxrate'], dataframe['density']),axis=1)
```
