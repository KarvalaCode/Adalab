```python
#Relación entre puesto de trabajo y satisfaccion de ambiente de trabajo
sns.set(style="dark")
sns.set_theme(style="dark")
plt.style.use('dark_background')


# Función para seleccionar aleatoriamente 50 empleados de cada grupo
def select_random_employees(group):
    return group.sample(min(len(group), 50)) 

# Filtrar aleatoriamente 50 empleados de cada departamento
filtered_df = df.groupby('job_role').apply(select_random_employees).reset_index(drop=True)

plt.figure(figsize=(12, 6))
sns.countplot(x='job_role', hue='environment_satisfaction', data=filtered_df, palette='flare', edgecolor='fuchsia')

# Etiquetas y título
plt.xlabel('Departamento')
plt.ylabel('Número de Empleados')
plt.title('Relación entre la satisfacción del ambiente de trabajo y puesto de trabajo')

plt.legend(title='Satisfacción amb. de trabajo')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#aqui he sacado la media pero es menos divertido
sns.set(style="dark")
sns.set_theme(style="dark")
plt.style.use('dark_background')


# Calcular la media de satisfacción del ambiente de trabajo para cada puesto de trabajo
mean_satisfaction = df.groupby('job_role')['environment_satisfaction'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='job_role', y='environment_satisfaction', data=mean_satisfaction, palette='flare', edgecolor='fuchsia')

# Etiquetas y título
plt.xlabel('Puesto de trabajo', fontsize=10)
plt.ylabel('Media de Satisfacción del Ambiente', fontsize=10)
plt.title('Media de Satisfacción del Ambiente por Puesto de Trabajo', fontsize=15)

plt.xticks(rotation=45, ha='right')

plt.ylim(2, None)
plt.tight_layout()
plt.show()
```

```python
#Relación entre la implicacion en el trabajo y porcentaje de aumento salarial
sns.set(style="dark")
sns.set_theme(style="dark")
plt.style.use('dark_background')

# Calcular la media de satisfacción del ambiente de trabajo para cada puesto de trabajo
mean_involvement = df.groupby('percent_salary_hike')['job_involvement'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='percent_salary_hike', y='job_involvement', data=mean_involvement, palette='flare', edgecolor='fuchsia')

# Etiquetas y título
plt.xlabel('Porcentaje de aumento salarial')
plt.ylabel('Media implicacion en el trabajo')
plt.title('Media de implicacion en el trabajo por porcentaje de aumento salarial')

plt.xticks(rotation=45, ha='right')

plt.ylim(2, None)
plt.tight_layout()
plt.show()
```
