```python
# List of columns to be renamed
columns_to_rename = ['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']

# Dictionary for renaming columns
rename_dict = {col: f"{col}_population" for col in columns_to_rename}

# Rename columns using the dictionary
df_population.rename(columns=rename_dict, inplace=True)
```

```python
# List of columns to be renamed
columns_to_rename = ['2013', '2014', '2015', '2016', '2017', '2018', '2019',
       '2020', '2021']

# Dictionary for renaming columns
rename_dict = {col: f"{col}_deaths" for col in columns_to_rename}

# Rename columns using the dictionary
df_deaths.rename(columns=rename_dict, inplace=True)
```

```python
df_death_population = df_deaths.merge(df_population, how='inner', on='country')
```

```python
df_death_population.drop('2022_population', axis=1, inplace=True)
```

```python
columns_rename = ['2013', '2014', '2015', '2016', '2017', '2018', '2019',
       '2020', '2021']
```

```python
for col in columns_rename:
    df_death_population[f"{col}_tasa"] = df_death_population[f"{col}_deaths"]/df_death_population[f"{col}_population"]
df_death_population    
```
